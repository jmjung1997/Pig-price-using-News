```python
import pandas as pd 
```

```python
df_sentiment=pd.read_csv('../sentimentResult_v3_all.csv',encoding='utf-8')
df_original=pd.read_csv('../../data/final_datasetv3.csv',encoding='utf-8')
```

```python
df_18=pd.read_excel('../../원본데이터/18년도 가격데이터.xlsx')
df_19=pd.read_excel('../../원본데이터/19년도 가격데이터.xlsx')
df_20=pd.read_excel('../../원본데이터/20년도 가격데이터.xlsx')
df_21=pd.read_excel('../../원본데이터/21년도 가격데이터.xlsx')
```

```python
df_18.drop(0,axis=0,inplace=True)
df_19.drop(0,axis=0,inplace=True)
df_20.drop(0,axis=0,inplace=True)
df_21.drop(0,axis=0,inplace=True)
```

```python
df_sentiment.tail()
```

### 감성 negative-> -1, positive->1, neural->0

```python
df_sentiment['sentiment'] = df_sentiment['sentiment'].replace('negative', -1)
df_sentiment['sentiment'] = df_sentiment['sentiment'].replace('neutral', 0)
df_sentiment['sentiment'] = df_sentiment['sentiment'].replace('positive', 1)
```

```python
df_sentiment.tail()
```

```python
grouped_df_sentiment = df_sentiment.groupby('week')['sentiment'].mean()
```

```python
print(grouped_df_sentiment)
```

```python
df_original['sentiment']=grouped_df_sentiment
```

```python
df_original.iloc[0:10]
```

### 날짜 기준으로 뉴스 감성점수 구하기

```python
grouped_df_original = df_original.groupby('datetime')['sentiment'].mean()
```

```python
df_original['day_sentiment'] = df_original.groupby(['datetime'])['sentiment'].transform('mean')
```

```python
df_original=df_original[['new_content','datetime','day_sentiment']]
```

```python
grouped_df = df_original.groupby('datetime')['new_content'].agg(lambda x: ' '.join(x)).reset_index()
```

```python
df_original.groupby('datetime')['new_content'].agg(lambda x: ' '.join(x))
```

```python
sentiment_final=pd.merge(grouped_df, df_original, on = 'datetime', how = 'left')
```

```python
sentiment_final=sentiment_final[['datetime','new_content_x','day_sentiment']]
```

```python
sentiment_final = sentiment_final.drop_duplicates('datetime', keep='last')
```

```python
sentiment_final.reset_index(drop=True,inplace=True)
```

```python
sentiment_final
```

### 가격 데이터 합치기

```python
df_price=pd.concat([df_18,df_19,df_20,df_21],axis = 0)
```

```python
df_price=df_price[['업체코드','업체명','도축일','도체번호','경락단가']]
```

```python
#도축일 datetime으로 변경
df_price['도축일']=pd.to_datetime(df_price['도축일'])
```

```python
# datetime으로 변경
df_original['datetime']=pd.to_datetime(df_original['datetime'])
```

### 일별 평균 데이터 구하기

```python
df_price_all_2column=df_price[['도축일',"경락단가"]]

df_price_all_2column['도축일']=pd.to_datetime(df_price_all_2column['도축일'])

df_price_all_2column.head()
```

```python
df_price_all_2column['경락단가_평균'] = df_price_all_2column.groupby(['도축일'])['경락단가'].transform('mean')

df_price_all_2column.head()

df_price_all_2column=df_price_all_2column.groupby(df_price_all_2column['도축일']).mean()

df_price_all_2column.reset_index(inplace=True)

new_df_price_all_2column=df_price_all_2column.iloc[238:,]

new_df_price_all_2column.reset_index(drop=True, inplace=True)
```

```python
new_df_price_all_2column
```

### 가격 데이터에 도축일 기준으로 merge하기

```python
# 기준열 이름이 다를 때
df_merged=pd.merge(new_df_price_all_2column, sentiment_final, left_on = '도축일', right_on = 'datetime', how = 'left')
```

```python
df_merged.fillna(method= 'ffill',inplace=True) # 뒤의 결측치를 앞의 값으로 채우기. 'pad'도 사용 가능하다.
```
