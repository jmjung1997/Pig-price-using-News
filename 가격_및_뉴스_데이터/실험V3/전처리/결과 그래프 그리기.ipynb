```python
import pandas as pd
```

```python
df=pd.read_excel("./결과표.xlsx")
```

```python
df
```

```python
LSTM_Value=df.iloc[-1,[2,6,10]]
```

```python
GRU_Value=df.iloc[-1,[4,8,12]]
```

```python
import matplotlib.pyplot as plt

# 데이터
categories = ['Day Sentiment', 'Wholesale Price', 'Wholesale price\n+\nDay Sentiment']
data1 = LSTM_Value
data2 = GRU_Value

# 그래프 너비
bar_width = 0.35

# X 축 위치 설정
x = range(len(categories))

# 첫 번째 막대 그래프
plt.bar(x, data1, width=bar_width,color='orange',hatch='xx' ,label='LSTM')

# 두 번째 막대 그래프 (오른쪽으로 옮김)
x2 = [i + bar_width for i in x]
plt.bar(x2, data2, width=bar_width,color='green' ,hatch='oo',label='GRU')

# X 축 레이블 설정
plt.xlabel('Three Cases')

# Y 축 레이블 설정
plt.ylabel('MAE')

# X 축 레이블 위치 설정
plt.xticks([i + bar_width / 2 for i in x], categories)

# 범례 표시
plt.legend()

# 그래프 제목 설정
plt.title('Average MAE Value in 3 cases')

# 그래프 표시
plt.show()
```
