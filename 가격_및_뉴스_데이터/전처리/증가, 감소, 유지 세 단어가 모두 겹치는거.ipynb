```python
data_1=data[data['label']==1]
data_2=data[data['label']==2]
data_0=data[data['label']==0]
```

```python
data_1_list=data_1['news'].values.tolist()
data_2_list=data_2['news'].values.tolist()
data_0_list=data_0['news'].values.tolist()
```

```python
vocab_1 = list(set(doc for day in data_1_list for doc in day))
vocab_2 = list(set(doc2 for day2 in data_2_list for doc2 in day2))
vocab_0 = list(set(doc0 for day0 in data_0_list for doc0 in day0))
```

```python
vocab=vocab_1+vocab_2+vocab_0
```

```python

# 중복되는 값을 저장할 빈 리스트 생성
result = []

# 리스트의 모든 요소를 순회하면서 중복된 값을 찾는다.
for i in vocab_0:
    if i in vocab_1 and i in vocab_2:
        result.append(i)
for j in vocab_1:
    if j in vocab_0 and j in vocab_2:
        result.append(j)
for k in vocab_2:
    if k in vocab_1 and k in vocab_0:
        result.append(k)
# 결과 출력
print(len(result))
```
