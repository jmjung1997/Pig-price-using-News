```python
# riss.kr 에서 특정 키워드로 논문 / 학술 자료 검색하기

#Step 1. 필요한 모듈을 로딩합니다
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time          

#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
print("=" *100)
print(" 이 크롤러는 경락단가 웹크롤러입니다.")
print("=" *100)
query_txt = input('1.시잣하는 날짜는 어떻게 됩니까?ex:2018-01-03 : ')
query_txt2 = input('2. 끝나는 날짜는 어떻게 됩니까?ex:2021-12-31 : ')


##Step 3. 수집된 데이터를 저장할 파일 이름 입력받기 
#ft_name = input('2.결과를 저장할 txt형식의 파일명을 쓰세요(예: c:\\py_temp\\riss.txt): ')
#fc_name = input('3.결과를 저장할 csv형식의 파일명을 쓰세요(예: c:\\py_temp\\riss.csv): ')
#fx_name = input('4.결과를 저장할 xls형식의 파일명을 쓰세요(예: c:\\py_temp\\riss.xls): ')

#Step 4. 크롬 드라이버 설정 및 웹 페이지 열기
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)

url = 'https://www.ekapepia.com/priceStat/distrPricePork.do'
driver.get(url)
time.sleep(2)
driver.maximize_window()

#Step 5. 날짜 선택란 선택
#시작날짜 선택하기

driver.find_element(By.ID,"day").click() 

element = driver.find_element(By.ID,'searchStartDate')
driver.find_element(By.ID,'searchStartDate').click()

#년도 선택
driver.find_element(By.CLASS_NAME,"ui-datepicker-year").click()
select = Select(driver.find_element(By.CLASS_NAME,"ui-datepicker-year"))                
select.select_by_visible_text('2018')

#월 선택
driver.find_element(By.CLASS_NAME,"ui-datepicker-month").click()
select = Select(driver.find_element(By.CLASS_NAME,"ui-datepicker-month"))                
select.select_by_visible_text('1월')

#일 선택
driver.find_element(By.XPATH,"//*[@id=\"ui-datepicker-div\"]/table/tbody/tr[1]/td[4]/a").click()


#끝날짜 선택하기

element = driver.find_element(By.ID,'searchEndDate')
driver.find_element(By.ID,'searchEndDate').click()

#년도 선택
driver.find_element(By.CLASS_NAME,"ui-datepicker-year").click()
select = Select(driver.find_element(By.CLASS_NAME,"ui-datepicker-year"))                
select.select_by_visible_text('2021')

#월 선택
driver.find_element(By.CLASS_NAME,"ui-datepicker-month").click()
select = Select(driver.find_element(By.CLASS_NAME,"ui-datepicker-month"))                
select.select_by_visible_text('12월')

#일 선택
driver.find_element(By.XPATH,"//*[@id=\"ui-datepicker-div\"]/table/tbody/tr[5]/td[6]/a").click()


#검색버튼 클릭
driver.find_element(By.ID,"ipt_search").click()

time.sleep(5)
```

```python
#Step 7.Beautiful Soup 로 본문 내용만 추출하기
from bs4 import BeautifulSoup
html_1 = driver.page_source
soup_1 = BeautifulSoup(html_1, 'html.parser')



#td_elements = soup_1.find('tbody').find_all('tr')

# Check if there is at least a second td element
#if len(td_elements) > 1:
 #   second_td = td_elements[1]
  #  print(second_td)
#else:
 #   print("There is no second td element.")

    
    
# Find all tr elements
tr_elements = soup_1.find('tbody').find_all('tr')

#Step 데이터 저장리스트
price = [ ]        #가격
date = [ ]         #날짜

# Iterate over each tr element
for tr in tr_elements:
    # Find all td elements within the current tr element
    th_elements=tr.find('th')
    print(th_elements.get_text())
    date.append(th_elements.get_text())
    td_elements = tr.find_all('td')
    # Iterate over each td element
    second_td = td_elements[1]
    price_tag = second_td.find(class_='mr5')

    price.append(price_tag.get_text())
```
