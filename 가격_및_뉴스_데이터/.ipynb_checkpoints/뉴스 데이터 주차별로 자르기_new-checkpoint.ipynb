### stem_list3 리스트 txt 파일로 부터 로드한 후 단어 빈도수 세기

```python
for i in range(len(stem_list3)):
    file = open('./stem/stem_list3_증감겹치는단어{}.txt'.format(i), 'r',encoding='UTF-8')
    text = file.read() 
    wordList = text.split()

    wordCount = {} 

    for word in wordList:

        # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴

        wordCount[word] = wordCount.get(word, 0) + 1 

        keys = sorted(wordCount.keys())

    #단어와 그에 대한 단어갯수를 담을 리스트 만들기
    word_list=[]
    wordCountlist=[]

    for word in keys:
        word_list.append(word)
        wordCountlist.append(wordCount[word])

    df_final=pd.DataFrame({'Word':word_list,'Count':wordCountlist})
    df_final_sort=df_final.sort_values(by=['Count'], ascending=False)
    df_final_sort.to_excel('./WordCount/word_count증감중복제거_{}.xlsx'.format(i))
```
